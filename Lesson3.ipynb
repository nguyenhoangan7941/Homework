```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import skfuzzy as fuzz
import seaborn as sns
dataset=pd.read_csv('/content/salary_Data.csv')
x=dataset.drop(['Salary'],axis=True)
y=dataset['Salary']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
sns.pairplot(dataset)
```

```python
plt.plot(x_train,y_train,'r.',label='train_data')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test).round(1)
cal=pd.DataFrame(np.c_[y_test,y_pred],columns=['Luong goc','Luong tien doan'])
cal
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,model.predict(x_train),color='blue')
plt.title('Salary vs Experience (Train Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show
print(model.coef_)
```

```python
import pandas as pd
data=pd.read_csv('/content/mtcars.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation=data.corr()
print(correlation)
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
sns.pairplot(data)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=[9,9],color='black')
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
model=LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg']) #Ham sklearn nho dung X in hoa
print('He so',model.coef_)
print('Giao diem',model.intercept_)
#ket qua y=-5,344x+37,285
```

```python
model.score(X=pd.DataFrame(data['wt']),y=data['mpg'])
#Do chinh xac
```

```python
#Ve duong hoi quy
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='blue',xlim=(0,7))
y_pred=model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='red')
```

```python
model1=LinearRegression()
pred=pd.DataFrame([data['wt'],data['wt']**2]).T
model1.fit(X=pred,y=data['mpg'])
print('He so',model1.coef_)
print('Giao diem',model1.intercept_)
print('Do chinh xac',model1.score(X=pred,y=data['mpg']))
#ket qua y=-13,38x+1,71x binh phuong + 49,93
```

```python
import pandas as pd
poly_range=np.arange(1.5,5.5,0.1)
poly_pred=pd.DataFrame([poly_range,poly_range**2]).T #tao cot du lieu, 1 cot x, 1 cot x binh phuong
y_pred=model1.predict(X=poly_pred)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(poly_range,y_pred,color='blue')
```

```python
model2=LinearRegression()
pred=pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3]).T
model2.fit(X=pred,y=data['mpg'])
print('He so',model2.coef_)
print('Giao diem',model2.intercept_)
print('Do chinh xac',model2.score(X=pred,y=data['mpg']))
```

```python
model3=LinearRegression()
pred=pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3,data['wt']**4]).T
model3.fit(X=pred,y=data['mpg'])
print('He so',model3.coef_)
print('Giao diem',model3.intercept_)
print('Do chinh xac',model3.score(X=pred,y=data['mpg']))
```

```python
model10=LinearRegression()
pred=pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3,data['wt']**4,data['wt']**5,data['wt']**6,
                   data['wt']**7,data['wt']**8,data['wt']**9,data['wt']**10]).T
model10.fit(X=pred,y=data['mpg'])
print('He so',model10.coef_)
print('Giao diem',model10.intercept_)
print('Do chinh xac',model10.score(X=pred,y=data['mpg']))
```

```python
import pandas as pd
poly_range=np.arange(1.5,5.5,0.1)
poly_pred=pd.DataFrame([poly_range,poly_range**2,poly_range**3,poly_range**4,poly_range**5,
                        poly_range**6,poly_range**7,poly_range**8,poly_range**9,poly_range**10]).T #tao cot du lieu, 1 cot x, 1 cot x binh phuong
y_pred=model10.predict(X=poly_pred)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(poly_range,y_pred,color='blue')
```
